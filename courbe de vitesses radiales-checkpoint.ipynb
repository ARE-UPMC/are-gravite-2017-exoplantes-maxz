```python
import numpy as np
t,v=np.loadtxt('vitesses radiales.txt',skiprows=2,usecols=(0,1),unpack=True)
%matplotlib inline
from pylab import *
figure(figsize=(10,5), dpi=80)
plot(t, v,color="blue", linewidth=1.5, linestyle="-")
xticks([ 2.5, 5, 7.5, 10],
       [ r'$2.5$', r'$5$', r'$7.5$', r'$+10$'])
yticks([-20, 0, +20],
       [r'$-20$', r'$0$', r'$+20$'])

ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.ylabel('v(km/s)')
plt.xlabel('t(jour)')
show()
```

Analyse d'une courbe de vitesses:

Cette courbe ressemble à une courbe sinus.
La maximum de la courbe est la vitesse radiale le plus grande ,cela correspond à l'éloignement de l'étoile.
Le minimum de la courbe est aussi la vitesse radiale le plus grande mais cela correspond à l'aprochement de l'étoile.
Lorsque la  vitesse radiale est le plus grande, le vecteur de la vitesse radiale est horizontale à la ligne droite d'observation.
lorsque la vitesse radiale est nulle,l'étoile est sur la ligne droite d'observation,et le vecteur de la vitesse radiale est perpendiculaire à la ligne droite d'observation.
